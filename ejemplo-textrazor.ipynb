La documentación para el [interfaz REST](http://www.textrazor.com/documentation_rest) y el [cliente en Python](http://www.textrazor.com/documentation_python). 

```python
import requests
URL = "http://api.textrazor.com"
apiKey = "a479239078b9438339c6d88330f3f3c90d14d31975e92d25336e1ec2"
text = """A $78-million state-of-the-art military vessel diverted to Alaska as a ferry hasn't worked out. 
Officials are desperate to be rid of it and Los Angeles might just be willing to take it."""
```

Existen varias funcionalidades disponibles, que se especifican como valores en el parámetro `extractors`.

- entities
- topics
- words
- dependency-trees
- relations
- entailments

Veamos algunos ejemplos:


### Entidades

Las [entities](http://www.textrazor.com/documentation_rest#entities) se enlazan con IDs e información sobre Wikipedia, FreeBase, etc., e incluyen índices de confianza.

```python
params = {"apiKey": apiKey, "extractors": "entities", "text": text}
r = requests.post(URL, data=params)
print r.text
```

### Temas

Los [topics](http://www.textrazor.com/documentation_rest#topics) extraen temas de Wikipedia, con índices de confianza.

```python
params = {"apiKey": apiKey, "extractors": "topics", "text": text}
r = requests.post(URL, data=params)
print r.text
```

### Palabras

Las [words](http://www.textrazor.com/documentation_rest#words) tokeniza y etiqueta las palabras, proporciona lemas y stems.

```python
params = {"apiKey": apiKey, "extractors": "words", "text": text}
r = requests.post(URL, data=params)
print r.text
```

### Árboles de dependencias

Los *dependecy-trees* 

```python
params = {"apiKey": apiKey, "extractors": "dependency-trees", "text": text}
r = requests.post(URL, data=params)
print r.text
```

```python
params = {"apiKey": apiKey, "extractors": "relations", "text": text}
r = requests.post(URL, data=params)
print r.text
```
